```python
import pandas as pd
from tabulate import tabulate
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
import joblib
```

```python
#Importation des données
path = 'creditcard.csv'
df = pd.read_csv(path)
df.isnull().sum()
df = df.dropna()
```

```python
#Data preprocessing
classes = df['Class'].value_counts()
X = df.iloc[:,:-1]
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)

clf = RandomForestClassifier()

param_grid = {
    'n_estimators':[800,400],
    'max_depth': [20],
}

# list_param = ['n_estimators', 'max_depth']

# for estimator in list_param: 
#     clf = RandomForestClassifier(estimator), 
    
```

```python
#Building the model
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, n_jobs=-1, scoring = 'f1')
print("Model")
grid_search.fit(X_train, y_train)
```

```python
joblib.dump(grid_search, "./random_forest.joblib")
```

```python
best_model = grid_search.best_estimator_
best_result = grid_search.best_score_
test_score = best_model.score(X_test, y_test)
print(grid_search.scorer_)
print("Meilleur modèle : ", best_model)
print("Score sur l'ensemble de test : ", best_result)
```

```python
test_score = best_model.score(X_test, y_test)
print(grid_search.scorer_)
print("Meilleur modèle : ", best_model)
print("Score sur l'ensemble de test : ", test_score)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

y_pred = grid_search.predict(X_test)
confusion_matrix = confusion_matrix(y_test,y_pred)

def confusion_matrix_function(classes, confusion_matrix):
    print("Confusion Matrix")
    # Créer les en-têtes de la table avec les classes prédites
    headers = [""] + [f"Predicted: {cls}" for cls in classes.index]

    # Créer les lignes de la table avec les valeurs de la matrice de confusion
    rows = [
        [f"Class: {classes.index[i]} - {classes[i]}"] + list(confusion_matrix[i])
        for i in range(len(classes))
    ]

    # Afficher la table
    print(tabulate(rows, headers=headers, tablefmt="grid"))

def metric_analyse(accuracy, precision, recall=None, f1_score=None):
  print('(accuray) ', accuracy)
  print('(precision) ', precision)
  print('(recall) ', recall)
  print('(f1_score) ', f1_score)
```
